```python
#Q8
# Initialize lists to store results for CSMA and CSMA/CD
csma_success_rate_list = []
csma_idle_rate_list = []
csma_collision_rate_list = []

csma_cd_success_rate_list = []
csma_cd_idle_rate_list = []
csma_cd_collision_rate_list = []

# Define parameters
link_delay_list = [0, 1, 2, 3]
packet_size_list = [7, 5, 3, 1]
n = 10

# Loop over different link delays
for link_delay, packet_size in zip(link_delay_list, packet_size_list):
    # Initialize cumulative results for CSMA and CSMA/CD
    csma_success_rate = 0
    csma_idle_rate = 0
    csma_collision_rate = 0
    
    csma_cd_success_rate = 0
    csma_cd_idle_rate = 0
    csma_cd_collision_rate = 0
    
    # Run simulations for n iterations
    for _ in range(n):
        # Create problem instance with current link delay and packet size for CSMA
        csma_problem = Setting(link_delay=link_delay, packet_size=packet_size)
        # Create problem instance with current link delay and packet size for CSMA/CD
        csma_cd_problem = Setting(link_delay=link_delay, packet_size=packet_size)
        
        # Perform simulations for CSMA
        csma_result = protocols.csma(csma_problem)
        csma_success_rate += csma_result[0]
        csma_idle_rate += csma_result[1]
        csma_collision_rate += csma_result[2]
        
        # Perform simulations for CSMA/CD
        csma_cd_result = protocols.csma_cd(csma_cd_problem)
        csma_cd_success_rate += csma_cd_result[0]
        csma_cd_idle_rate += csma_cd_result[1]
        csma_cd_collision_rate += csma_cd_result[2]
    
    # Calculate averages over n iterations for CSMA and CSMA/CD
    csma_success_rate /= n
    csma_idle_rate /= n
    csma_collision_rate /= n
    
    csma_cd_success_rate /= n
    csma_cd_idle_rate /= n
    csma_cd_collision_rate /= n
    
    # Append results to lists
    csma_success_rate_list.append(csma_success_rate)
    csma_idle_rate_list.append(csma_idle_rate)
    csma_collision_rate_list.append(csma_collision_rate)
    
    csma_cd_success_rate_list.append(csma_cd_success_rate)
    csma_cd_idle_rate_list.append(csma_cd_idle_rate)
    csma_cd_collision_rate_list.append(csma_cd_collision_rate)
```

```python
plt.plot(link_delay_list, csma_success_rate_list, marker='^', label='csma', color='#00ff00')
plt.plot(link_delay_list, csma_cd_success_rate_list, marker='o', label='csma/cd', color='#0000ff')
plt.title("Success Rate vs. Link Delay")
plt.xlabel("Link Delay")
plt.ylabel("Success Rate")
plt.legend()
plt.show()

plt.plot(link_delay_list, csma_idle_rate_list, marker='^', label='csma', color='#00ff00')
plt.plot(link_delay_list, csma_cd_idle_rate_list, marker='o', label='csma/cd', color='#0000ff')
plt.title("Idle Rate vs. Link Delay")
plt.xlabel("Link Delay")
plt.ylabel("Idle Rate")
plt.legend()
plt.show()

plt.plot(link_delay_list, csma_collision_rate_list, marker='^', label='csma', color='#00ff00')
plt.plot(link_delay_list, csma_cd_collision_rate_list, marker='o', label='csma/cd', color='#0000ff')
plt.title("Collision Rate vs. Link Delay")
plt.xlabel("Link Delay")
plt.ylabel("Collision Rate")
plt.legend()
plt.show()
```
